## List

Nah, kita bisa menggunakan variabel bertipe list
serta perulangan for. Variabel nama, pelajaran, dan nilai merupakan variabel dengan tipe list. Namun, item pada variabel nama dan pelajaran merupakan tipe string. Sedangkan item pada variabel nilai merupakan tipe integer.

```python
nama = ['Sinubi','Budi','Ayu','Tono','Putri']
pelajaran = ["Pemrograman Python", "Pemrograman Python",
             "Pemrograman Python", "Machine Learning",
             "Machine Learning"]
nilai = [85, 80, 90, 88, 83]
```

Untuk menampilkan pesan, kita menggunakan perulangan for. Kita akan menampilkan pesan sebanyak jumlah siswa yang ada pada variabel nama yaitu 5 siswa. Di sini, kita menggunakan `len()` untuk mendapatkan jumlah siswa pada variabel nama. Sedangkan `range()` digunakan untuk membuat urutan index i dari 0 sampai dengan jumlah siswa – 1 yaitu 0, 1, 2, 3, dan 4. Dengan begitu, untuk setiap iterasi perulangan, kita dapat menampilkan item dengan index i dari list.

```python
for i in range(0, len(nama)):
  print(nama[i] + ' mendapat nilai ' + str(nilai[i]) +
        ' untuk pelajaran ' + pelajaran[i])
```

Kode Keseluruhan :

```python
nama = ['Sinubi','Budi','Ayu','Tono','Putri']
pelajaran = ["Pemrograman Python", "Pemrograman Python",
             "Pemrograman Python", "Machine Learning",
             "Machine Learning"]
nilai = [85, 80, 90, 88, 83]

for i in range(0, len(nama)):
  print(nama[i] + ' mendapat nilai ' + str(nilai[i]) +
        ' untuk pelajaran ' + pelajaran[i])
```

Sekarang mari kita perhatikan struktur data yang telah kita buat. Di sini, kita menggunakan 3 variabel bertipe list. Setiap list di sini dapat kita ilustrasikan sebagai satu kolom data.

```python
data = [{'nama':'Sinubi', 'pelajaran':'Pemrograman Python','nilai':85},
        {'nama':'Budi'  , 'pelajaran':'Pemrograman Python','nilai':80},
        {'nama':'Ayu'   , 'pelajaran':'Pemrograman Python','nilai':90},
        {'nama':'Tono'  , 'pelajaran':'Machine Learning'  ,'nilai':88},
        {'nama':'Putri' , 'pelajaran':'Machine Learning'  ,'nilai':83}]
```

Maka dari itu, untuk setiap iterasi perulangan,
kita dapat menampilkan item dari list yang

item dari list akan diwakili dengan variabel ‘baris’. Kemudian, kita menggunakan key ‘nama’, ‘pelajaran’, dan ‘nilai’ untuk menampilkan informasi yang kita inginkan.


```python
for baris in data:
  print(baris['nama'] + ' mendapat nilai ' + str(baris['nilai']) +
        ' untuk pelajaran ' + baris['pelajaran'])
```

## Pengkondisian

Selanjutnya, Sinubi ingin menambahkan informasi
nilai dalam huruf seperti berikut :

image.png

Untuk menampilkan nilai huruf, Sinubi menggunakan
pengkondisian if, else if, dan else. Jika kita hanya mempunyai 2 kemungkinan kondisi, maka kita cukup menggunakan if dan else saja. Nampun apabila kita mempunyai lebih dari 2 kemungkinan kondisi, maka kita menggunakan if, else if, dan else.

```python
nilai = 83
```

```python
nilai_huruf = ''

if nilai >= 85 :
  nilai_huruf = 'A'
elif (nilai >= 75) and (nilai < 85) :
  nilai_huruf = 'B'
elif (nilai >= 50) and (nilai < 75) :
  nilai_huruf = 'C'
else :
  nilai_huruf = 'D'
```

```python
nilai_huruf
```

Sekarang, mari kita tambahkan kode pengkondisian
ke bagian kode perulangan yang telah kita buat sebelumnya. Kemudian, tambahkan informasi nilai huruf pada fungsi print(). Perhatikan bahwa kita membutuhkan index i untuk mendapatkan nilai setiap siswa atau dengan kata lain item pada list nilai.

Kode Keseluruhan :

```python
nama = ['Sinubi','Budi','Ayu','Tono','Putri']
pelajaran = ["Pemrograman Python", "Pemrograman Python",
             "Pemrograman Python", "Machine Learning",
             "Machine Learning"]
nilai = [85, 80, 90, 88, 83]

for i in range(0, len(nama)):
  nilai_huruf = ''

  if nilai[i] >= 85 :
    nilai_huruf = 'A'
  elif (nilai[i] >= 75) and (nilai[i] < 85) :
    nilai_huruf = 'B'
  elif (nilai[i] >= 50) and (nilai[i] < 75) :
    nilai_huruf = 'C'
  else :
    nilai_huruf = 'D'

  print(nama[i] + ' mendapat nilai ' + str(nilai[i]) +
        ' (' + nilai_huruf + ')' +
        ' untuk pelajaran ' + pelajaran[i])
```

## Operator

Mari kita kembali ke kode pengkondisian berikut.
Di sini, kita menggunakan beberapa operator.

1.   Pertama adalah kelompok operator komparasi
yaitu >= dan <
2.   Berikutnya yang kedua adalah
kelompok operator logika yaitu and
3.   Dan terakhir adalah
kelompok operator pemberian nilai yaitu =

```python
nilai_huruf = ''

if nilai >= 85 :
  nilai_huruf = 'A'
elif (nilai >= 75) and (nilai < 85) :
  nilai_huruf = 'B'
elif (nilai >= 50) and (nilai < 75) :
  nilai_huruf = 'C'
else :
  nilai_huruf = 'D'
```
