## Fungsi

Fungsi adalah blok kode yang hanya akan dijalankan ketika dipanggil.  Sebuah Fungsi dapat memiliki masukan yang disebut sebagai argumen (biasanya disebut juga sebagai parameter). Selain itu, sebuah Fungsi juga dapat mengeluarkan/mengembalikan sebuah nilai dengan menggunakan perintah return.


image.png

Berikut adalah fungsi untuk mendapatkan nilai huruf :

```python
# Membuat Fungsi
def dapatkan_nilai_huruf(nilai_angka):
  if nilai_angka >= 85 :
    nilai_huruf = 'A'
  elif (nilai_angka >= 75) and (nilai_angka < 85) :
    nilai_huruf = 'B'
  elif (nilai_angka >= 50) and (nilai_angka < 75) :
    nilai_huruf = 'C'
  else :
    nilai_huruf = 'D'

  return nilai_huruf
```

Untuk membuat fungsi, kita membutuhkan kata kunci `def`. Setelah kata kunci `def`, sertakan nama fungsi yang diinginkan. Misalnya ‘dapatkan_nilai_huruf’. Fungsi dapat memiliki masukan. Argumen/parameter sebuah fungsi dapat kita definisikan di dalam tanda kurung setelah nama fungsi. Bagian selanjutkan merupakan kode pengkondisian yang merupakan blok kode dari fungsi yang kita buat. Untuk mengembalikan sebuah nilai, gunakan kata kunci return. Pada kasus ini, fungsi yang kita buat akan mengeluarkan ‘nilai_huruf’.

Sebuah fungsi bernama ‘dapatkan_nilai_huruf’ telah dibuat. Fungsi ini akan menerima masukan berupa nilai angka dan akan mengeluarkan nilai huruf dari nilai angka yang diberikan tersebut.

Mari kita jalankan fungsi bernama ‘dapatkan_nilai_huruf’ yang telah dibuat. Untuk menjalankan sebuah fungsi, panggilah fungsi tersebut dengan namanya dan sertakanlah argumen/parameter yang akan menjadi masukan bagi fungsi tesebut.


```python
# Memanggil Fungsi
dapatkan_nilai_huruf(77)
```

Sekarang, kode program yang telah kita buat menjadi lebih sederhana karena telah menggunakan fungsi ‘dapatkan_nilai_huruf’.

```python
nama = ['Sinubi','Budi','Ayu','Tono','Putri']
pelajaran = ["Pemrograman Python", "Pemrograman Python",
             "Pemrograman Python", "Machine Learning",
             "Machine Learning"]
nilai = [85, 80, 90, 88, 83]

for i in range(0, len(nama)):
  nilai_huruf = dapatkan_nilai_huruf(nilai[i])

  print(nama[i] + ' mendapat nilai ' + str(nilai[i]) +
        ' (' + nilai_huruf + ')' +
        ' untuk pelajaran ' + pelajaran[i])
```

## Scope


Ada satu hal yang harus kita perhatikan jika kita menggunakan fungsi yaitu scope. Mari perhatikan fungsi berikut :

```python
# Membuat Fungsi
def sebuah_fungsi():
  sebuah_variabel = 'ini di dalam fungsi'   # variabel lokal
  print(sebuah_variabel)
```

Pada fungsi yang bernama ‘sebuah_fungsi’,
terdapat sebuah variabel bertipe string
dengan nama ‘sebuah_variabel’. Variabel yang bernama ‘sebuah_variabel’ ini adalah variabel lokal. Artinya, variabel ini hanya berlaku/hidup di dalam fungsi tersebut.

Selain variabel lokal, adapula variabel global.
Variabel global ini berlaku/hidup diluar suatu fungsi.

Mari kita bahas contoh kode program berikut.

image.png

Pertama, kita membuat sebuah fungsi bernama ‘sebuah fungsi’. Pada fungsi tersebut, terdapat sebuah variabel lokal bernama ‘sebuah_variabel’. Pada saat dijalankan, fungsi tersebut akan menampilkan nilai dari variabel ‘sebuah_variabel’. Mari kita beralih ke Cell 2. Di sini, pertama-tama kita mendefinisikan sebuah variabel global yang bernama ‘sebuah variabel’. Walaupun memiliki nama yang sama yaitu ‘sebuah variabel’, variabel lokal di dalam fungsi merupakan entitas yang berbeda dengan variabel global di luar fungsi. Hal ini dapat terlihat ketika kita memanggil fungsi ‘sebuah_fungsi’ maka print menampilkan nilai ‘ini di dalam fungsi’. Namun setelahnya, print menampilkan
nilai ‘ini di luar fungsi’. Padahal, print sama-sama menampilkan nilai dari variabel ‘sebuah_variabel’. Perbedaan ini terjadi karena penggunaan variabel lokal dan variabel global.

```python
sebuah_variabel = 'ini di luar fungsi'   # variabel global

sebuah_fungsi() # memanggil fungsi

print(sebuah_variabel)
```

## Docstrings

Sinubi telah membuat fungsi bernama ‘dapatkan_nilai_huruf’. Tentu saja sebagai pembuat fungsi tersebut, Sinubi tahu kegunaan serta cara kerja dari fungsi ini. Namun belum tentu untuk programmer lain.
Bisa saja programmer lain tidak tahu kegunaan dari fungsi yang dibuat/ditulis. Maka dari itu, Sinubi dapat menambahkan informasi mengenai fungsi tersebut dengan menambahkan docstrings. Cukup tambahkan informasi tentang fungsi yang telah dibuat setelah mendefinisikan nama fungsi. Pesan informasi diapit dengan ”””.

```python
# Membuat Fungsi
def dapatkan_nilai_huruf(nilai_angka):
  """Fungsi untuk mengubah nilai angka menjadi nilai huruf.

    * A : nilai ≥ 85
    * B : 75 ≥ nilai < 85
    * C : 50 ≥ nilai < 75
    * D : nilai < 50
  """
  if nilai_angka >= 85 :
    nilai_huruf = 'A'
  elif (nilai_angka >= 75) and (nilai_angka < 85) :
    nilai_huruf = 'B'
  elif (nilai_angka >= 50) and (nilai_angka < 75) :
    nilai_huruf = 'C'
  else :
    nilai_huruf = 'D'

  return nilai_huruf
```

Kita kemudian bisa melihat informasi fungsi tersebut dengan perintah `help`.

```python
help(dapatkan_nilai_huruf)
```

Selain itu, informasi fungsi juga dapat dilihat dengan menuliskan nama fungsi tersebut diikuti dengan tanda tanya ‘?’.

```python
!dapatkan_nilai_huruf?
```

## Lambda

Sejauh ini Sinubi telah berhasil membuat sebuah fungsi berserta informasi mengenai fungsi tersebut sehingga orang lain tahu gambaran maupun cara kerja dari fungsi tersebut.

Sinubi kemudian mendapat kabar bahwa nilai siswa akan diberi bonus 5 poin akibat kesalahan soal ujian pada mata pelajaran “Pemrograman Python” dan “Machine Learning”.

Untuk mengatasi permasalahan ini, Sinubi menggunakan fungsi lambda. Lambda merupakan fungsi sederhana dengan syntax penulisan sebagai berikut :

image.png

Kita bisa menambahkan ke kode program sebelumnya sehingga menjadi seperti ini :

```python
nama = ['Sinubi','Budi','Ayu','Tono','Putri']
pelajaran = ["Pemrograman Python", "Pemrograman Python",
             "Pemrograman Python", "Machine Learning",
             "Machine Learning"]
nilai = [85, 80, 90, 88, 83]

bonus = lambda x : x + 5

for i in range(0, len(nama)):
  nilai_huruf = dapatkan_nilai_huruf(bonus(nilai[i]))

  print(nama[i] + ' mendapat nilai ' + str(bonus(nilai[i])) +
        ' (' + nilai_huruf + ')' +
        ' untuk pelajaran ' + pelajaran[i])
```

Fungsi lambda yang telah dibuat diberi nama ‘bonus’. Fungsi lambda ‘bonus’ dipanggil pada saat pengkonversian nilai angka menjadi huruf serta pada saat menampilkan nilai siswa dengan fungsi print().

yang telah dia buat. Selanjutnya, Sinubi berencana untuk menyimpan pesan keluaran program rapor siswa ke dalam file teks.

Berikut kode program untuk menyimpan pesan keluaran program rapor siswa ke dalam file teks :

```python
nama = ['Sinubi','Budi','Ayu','Tono','Putri']
pelajaran = ["Pemrograman Python", "Pemrograman Python",
             "Pemrograman Python", "Machine Learning",
             "Machine Learning"]
nilai = [85, 80, 90, 88, 83]

bonus = lambda x : x + 5

alamat_file = '/content/rapor_siswa.txt'
file = open(alamat_file,'w')

for i in range(0, len(nama)):
  nilai_huruf = dapatkan_nilai_huruf(bonus(nilai[i]))

  pesan = nama[i] + ' mendapat nilai ' + str(bonus(nilai[i])) + \
          ' (' + nilai_huruf + ')' + \
          ' untuk pelajaran ' + pelajaran[i]

  print(pesan)
  file.write(pesan + '\n')

file.close()
```

Pertama, file teks dibuka dahulu oleh program. Kedua, file teks ditulis oleh program dengan pesan keluaran berupa informasi nilai siswa. Ketiga, file teks ditutup oleh program.

## DataFrame

DataFrame sederhananya merupakan data terstruktur seperti tabel yang terdiri dari kolom dan baris.


image.png

Untuk membuat dataframe, gunakan class ‘DataFrame’ seperti contoh berikut. Kemudian, definisikan nama kolom untuk dataframe yang kita buat. Untuk setiap kolom, tujukan list yang berisi data. Sesuaikan nama kolom dengan list yang cocok.

```python
nama = ['Sinubi','Budi','Ayu','Tono','Putri']
pelajaran = ["Pemrograman Python", "Pemrograman Python",
             "Pemrograman Python", "Machine Learning",
             "Machine Learning"]
nilai = [85, 80, 90, 88, 83]

df = pd.DataFrame({'nama siswa'     : nama,
                   'mata pelajaran' : pelajaran,
                   'nilai siswa'    : nilai})
```

Kita telah berhasil membuat dataframe yang berisi data siswa. Untuk melihat informasi mengenai dataframe yang telah dibuat, gunakan method ‘info’.


```python
df.info()
```

Selain itu, untuk melihat 5 data pertama pada dataframe yang telah dibuat, gunakan method ‘head’.


```python
df.head()
```

Sebaliknya, untuk melihat 5 data terakhir pada dataframe, gunakan method ‘tail’.

```python
df.tail()
```


Saat ini, data siswa yang kita miliki sudah berbentuk dataframe. Kita dapat melakukan export data kita menjadi sebuah file .csv dengan method ‘to_csv’.

```python
df.to_csv("data siswa.csv")
```


Selain file .csv, kita juga dapat melakukan export data menjadi file excel. Kita dapat melakukan export data kita menjadi sebuah file .xlsx dengan method ‘to_excel’.

```python
df.to_excel("data siswa.xlsx")
```
